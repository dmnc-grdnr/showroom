```python
import matplotlib.pyplot as plt
import pandas as pd
from common import rename_feature
```

```python
org_statistics = pd.read_csv("../data/netflow_statistics/org_Benign_statistics.csv", index_col=[0])
org_statistics
```

```python
sample_statistics = pd.read_csv("../data/netflow_statistics/sample_Benign_statistics.csv", index_col=[0])
sample_statistics
```

```python
for column in ['IPV4_DST_ADDR', 'IPV4_SRC_ADDR', 'DNS_QUERY_ID', 'Label', 'Attack', 'Dataset']:
    if org_statistics.get(column) is not None:
        org_statistics = org_statistics.drop(column, axis=1)
    if sample_statistics.get(column) is not None:
        sample_statistics = sample_statistics.drop(column, axis=1)
```

```python
org_statistics.columns = [rename_feature(name) for name in org_statistics.columns]
sample_statistics.columns = [rename_feature(name) for name in sample_statistics.columns]
```

```python
# see https://stackoverflow.com/questions/66563727/how-to-make-boxplot-using-matplotlib-when-i-already-have-all-the-information
def update(d_statistic, origin):
    IQRResult = d_statistic["75%"] - d_statistic["25%"]
    # Reasonable Upper Boundary (RUB)
    RUB = d_statistic["75%"] + (1.5 * IQRResult)
    # Reasonable Lower Boundary (RLB)
    RLB = d_statistic["25%"] - (1.5 * IQRResult)
    d_statistic["label"] = origin
    d_statistic["q1"] = d_statistic["25%"]
    d_statistic["med"] = d_statistic["50%"]
    d_statistic["q3"] = d_statistic["75%"]
    #d_statistic["whislo"]= d_statistic["min"]
    d_statistic["whislo"] = RLB  # required
    #d_statistic["whishi"]= d_statistic["max"]
    d_statistic["whishi"] = RUB  # required
    d_statistic["fliers"] = []#[d_statistic["min"], d_statistic["max"]]  # required if showfliers=True
    return d_statistic
```

```python
column = "DST_TO_SRC_AVG_TP"
d_org = ({k:v for k,v in zip(org_statistics.index, list(org_statistics[column]))})
d_org = update(d_org, "org")
d_sample = ({k:v for k,v in zip(sample_statistics.index, list(sample_statistics[column]))})
d_sample = update(d_sample, "sample")

fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(6, 6), sharey=True)#sharey -> same axis for all subplots
axes.bxp([d_sample])
axes.set_title(column, fontsize=10)
plt.show()
```

```python
def is_similar(sample_value, org_value):
    delta = org_value * 0.1
    return org_value - delta <= sample_value <= org_value + delta
```

```python
fig, axes = plt.subplots(nrows=5, ncols=4, figsize=(16, 20), sharey=False)
stats = []
i = j = 0
for column in sorted(org_statistics.columns):
    d_org = ({k:v for k,v in zip(org_statistics.index, list(org_statistics[column]))})
    d_org = update(d_org, "org")
    d_sample = ({k:v for k,v in zip(sample_statistics.index, list(sample_statistics[column]))})
    d_sample = update(d_sample, "sample")
    if not is_similar(d_sample["med"], d_org["med"]):
        a = axes[i][j]
        a.set_title(column, fontsize=10)
        a.bxp([d_org, d_sample])
        j += 1
        if j > 3:
            j = 0
            i += 1

plt.show()
```
